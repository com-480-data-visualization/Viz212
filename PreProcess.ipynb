## categorical attributes : Encoding

It seems better for later to encode the "Interests" attribute. One-hot-encoding is suitable for later vizualisations using bar charts for example.

```python
from sklearn.preprocessing import MultiLabelBinarizer
import pandas as pd

dataset["Interests"] = dataset["Interests"].apply(eval) if isinstance(dataset["Interests"].iloc[0], str) else dataset["Interests"]

# Normalize interests: strip spaces and convert to lowercase
dataset["Interests"] = dataset["Interests"].apply(lambda x: [interest.strip().lower() for interest in x])

# Apply MultiLabelBinarizer
mlb = MultiLabelBinarizer()
encoded_interests = pd.DataFrame(mlb.fit_transform(dataset["Interests"]), columns=mlb.classes_)

# Merge back with original dataset
df = dataset.join(encoded_interests).drop(columns=["Interests"])
```

```python
df['Education Level'].unique()
```

Map 
- "Gender" with {0: Male, 1: Female} 
- "Children" with {0: No, 1:Yes, 2: Maybe}
- "Occupation" with {'High School' : 'HS',"Master's Degree" : "MSc", "Bachelor's Degree":"BSc", 'Ph.D.' : "PhD" }

```python
df['Children'].unique()
```

```python
#Map Gender to numerical values
df['Gender'] = df['Gender'].map({"Male" : 0, "Female": 1})

#Map Children to numerical values
df['Children'] = df['Children'].map({"No": 0, "Yes": 1, "Maybe": 2})

#Map Education Level
df['Education Level'] = df['Education Level'].map({'High School' : 'HS',"Master's Degree" : "MSc", "Bachelor's Degree":"BSc", 'Ph.D.' : "PhD" })
```

```python
df.to_csv("dataset_preprocessed.csv", index=False)
```
